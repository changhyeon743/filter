```python
import os
import pandas as pd
```

```python
good = pd.read_csv("good.csv", header=None)
bad = pd.read_csv("bad.csv", header = None)
```

```python
good.head()
```

```python
bad.head()
```

```python
print(len(good),len(bad))
```

```python
g = pd.DataFrame(good)
g["label"] = 1
g.head()
```

```python
b = pd.DataFrame(bad)
b["label"] = 2
b.head()
```

```python
data = g.append(b)
len(data)
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
train,test = train_test_split(data, test_size=0.33, random_state= 42)
```

```python
train["label"] = pd.Categorical(train["label"])
```

- 위 코드는 훈련데이터에 Categorical(연구필요)을 이용해 분류를 해주는 것.
- 아래엔 이미 0.66배의 훈련데이터가 남았다.
- 나머지 0.33은 실험데이터.

```python
train.groupby("label").count()
```

```python
test.groupby("label").count()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from konlpy.tag import Twitter
```

```python
def get_noun(text):
    tokenizer = Twitter()
    nouns = tokenizer.nouns(text)
    return [n for n in nouns]
```

noun -> 명사추출

```python
cv = CountVectorizer(tokenizer=get_noun)
```

```python
tdm = cv.fit_transform(train[0])
```

```python
cv.vocabulary_
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.linear_model import SGDClassifier
from sklearn.pipeline import Pipeline

text_clf_svm = Pipeline([('vect', CountVectorizer(tokenizer=get_noun)),
                         ('tfidf', TfidfTransformer()),
                         ('clf-svm', SGDClassifier(loss='hinge', penalty='l2', alpha=1e-3, random_state=42))])

text_clf_svm = text_clf_svm.fit(train[0], train["label"])
```

```python
#TypeError: __init__() got an unexpected keyword argument 'n_iter'
#n_iter = 5
```

```python
predicted_svm = text_clf_svm.predict(test[0])
np.mean(predicted_svm == test["label"])
```

```python
text_clf_svm.predict(["오졌다","사랑해","배움","기뻐","아아"])
```

```python
# save the model to disk
import pickle

filename = 'model.sav'
pickle.dump(text_clf_svm, open(filename, 'wb'))

# some time later...
```

```python

# load the model from disk
loaded_model = pickle.load(open(filename, 'rb'))
print(loaded_model.predict(["영감"]))
```
